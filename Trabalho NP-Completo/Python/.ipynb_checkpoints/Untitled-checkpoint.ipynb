```python
import networkx as nx
import matplotlib.pyplot as plt

def is_valid_graph(graph):
    """
    Verifica se o dicionário fornecido representa um grafo válido.

    Argumentos:
    graph -- O grafo representado como um dicionário de adjacência.

    Retorna:
    Uma string indicando o motivo da invalidade se o dicionário não representa um grafo válido,
    ou None se o grafo for válido.
    """
    if not isinstance(graph, dict):
        return "O argumento fornecido não é um dicionário."

    for vertex, neighbors in graph.items():
        if not isinstance(neighbors, list):
            return f"Os vizinhos de '{vertex}' não são representados como uma lista."
        for neighbor in neighbors:
            if neighbor not in graph:
                return f"O vértice '{neighbor}' em 'graph['{vertex}']' não está presente como um vértice válido."

    return None

def is_independent_set(graph, vertices):
    """
    Verifica se um conjunto de vértices é um conjunto independente em um grafo.

    Argumentos:
    graph -- O grafo representado como um dicionário de adjacência.
    vertices -- O conjunto de vértices a ser verificado.

    Retorna:
    True se o conjunto de vértices é um conjunto independente, False caso contrário.
    """
    for u in vertices:
        for v in vertices:
            if u != v and v in graph[u]:
                return False
    return True

def draw_graph(graph, independent_set):
    G = nx.Graph(graph)
    pos = nx.spring_layout(G)
    
    nx.draw_networkx_nodes(G, pos, node_size=700)
    nx.draw_networkx_labels(G, pos)
    nx.draw_networkx_edges(G, pos)
    
    nx.draw_networkx_nodes(G, pos, nodelist=independent_set, node_color='r', node_size=700)
    
    plt.title('Grafo com conjunto independente destacado')
    plt.axis('off')
    plt.show()

# Exemplo de uso
I = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'K'],
    'K': ['C'],
    'D': ['B', 'C']
}

valid = is_valid_graph(I)

if valid == None:

    graph = I

    independent_set = ['A', 'D']
    result = is_independent_set(graph, independent_set)
    
    print(result)  # Saída: True
    
    if result:
        draw_graph(graph, independent_set)

else:
    print(f'Instancia I invalida: {valid}')
```

```python
def reduce_A_to_B(expression):
    graph = {}

    expression = expression.replace('(', '')
    expression = expression.replace(')', '')
    
    # Separando cláusulas
    clauses = expression.split(' AND ')
    
    # Criando vértices para cada variável em cada cláusula
    for clause in clauses:
        literals = clause.split(' OR ')
        for literal in literals:
            variable = literal.strip('-')
            if variable not in graph:
                graph[variable] = []
    
    # Criando arestas entre variáveis negadas na mesma cláusula
    for clause in clauses:
        literals = clause.split(' OR ')
        for i in range(len(literals)):
            for j in range(i + 1, len(literals)):
                literal1 = literals[i]
                literal2 = literals[j]
                if literal1[0] == '-' and literal2[0] == '-':
                    v1 = literal1.strip('-')
                    v2 = literal2.strip('-')
                    graph[v1].append(v2)
                    graph[v2].append(v1)
    
    return graph

# Exemplo de uso
expression = "(A OR B OR -C) AND (-A OR -B OR C) AND (A OR -B OR C)"
graph = reduce_A_to_B(expression)
print(graph)

draw_graph(graph, [])
```

```python
def instancia_A(I):
    # Para o problema A', a instância I é a própria entrada, então simplesmente retornamos I
    return I

def f(I):
    # Redução da instância do problema A' (3CNF-SAT) para o problema B' (Independent Set)
    # Vamos realizar a redução da seguinte forma:
    # - Para cada cláusula C na fórmula 3CNF-SAT, criamos um vértice no grafo do Independent Set.
    # - Para cada variável V em C e sua negação ¬V, conectamos os vértices correspondentes no grafo com uma aresta.
    
    # Primeiro, dividimos a instância I em cláusulas
    clauses = I.split(" ∧ ")
    
    # Vamos usar um contador para nomear os vértices do Independent Set
    vertex_count = 1
    graph = {}  # Inicializamos o grafo do Independent Set
    
    # Iteramos sobre cada cláusula
    for clause in clauses:
        variables = clause.split(" ∨ ")
        
        # Para cada variável em uma cláusula, criamos um vértice no grafo
        for var in variables:
            # Se a variável for negada, removemos o caractere ¬
            if var[0] == "¬":
                vertex = "¬V" + str(vertex_count)
            else:
                vertex = "V" + str(vertex_count)
            vertex_count += 1
            
            # Adicionamos o vértice ao grafo
            graph[vertex] = []
            
            # Conectamos o vértice a todos os vértices já existentes na cláusula
            for existing_vertex in graph.keys():
                if existing_vertex != vertex:
                    graph[vertex].append(existing_vertex)
                    graph[existing_vertex].append(vertex)
    
    # Retornamos o grafo como a instância reduzida do problema B'
    return graph

def algoritmo_B(instancia_f):
    # Agora, para resolver o problema B' (Independent Set), vamos encontrar o conjunto independente máximo no grafo.
    # Vamos utilizar um algoritmo de busca exaustiva para isso.
    
    max_independent_set = set()  # Inicializamos o conjunto independente máximo
    
    # Iteramos sobre todos os vértices do grafo
    for vertex in instancia_f:
        current_set = {vertex}  # Começamos com um conjunto contendo apenas o vértice atual
        
        # Verificamos quais vértices adjacentes não estão no conjunto e os adicionamos
        for neighbor in instancia_f[vertex]:
            if neighbor not in current_set:
                current_set.add(neighbor)
        
        # Verificamos se o conjunto atual é maior do que o conjunto independente máximo encontrado até agora
        if len(current_set) > len(max_independent_set):
            max_independent_set = current_set
    
    # Retornamos o conjunto independente máximo encontrado
    return max_independent_set

def h(S):
    # Como S é o conjunto independente máximo no grafo do Independent Set, podemos simplesmente retorná-lo como resposta para o problema A'
    return S

# Exemplo de uso das funções
I = "(A ∨ B ∨ ¬C) ∧ (¬A ∨ ¬B ∨ ¬C) ∧ (¬A ∨ ¬B ∨ C)"  # instância do problema A'
print(f"instância do problema A': {I}")
instancia_f = f(I)  # Redução da instância do problema A' para o problema B'
print(f"Redução da instância do problema A' para o problema B': \n{instancia_f}")
S = algoritmo_B(instancia_f)  # Resolução do problema B' para a instância f(I)

import networkx as nx
import matplotlib.pyplot as plt

# Função para desenhar o grafo com destaque para o conjunto independente máximo
def draw_graph_with_independent_set(graph, independent_set):
    G = nx.Graph(graph)
    pos = nx.spring_layout(G)  # Layout do grafo
    
    # Desenhar os vértices
    nx.draw_networkx_nodes(G, pos, node_size=700)
    
    # Desenhar as arestas
    nx.draw_networkx_edges(G, pos)
    
    # Desenhar as etiquetas dos vértices
    nx.draw_networkx_labels(G, pos)
    
    # Destacar o conjunto independente máximo
    nx.draw_networkx_nodes(G, pos, nodelist=independent_set, node_color='r', node_size=700)
    
    plt.title("Grafo com Conjunto Independente Máximo")
    plt.axis('off')
    plt.show()


if S is not None:
    resposta_A = h(S)  # Converte a solução do problema B' de volta para o problema A'
    print("Resposta para o problema A':", resposta_A)

    # Exemplo de uso
    graph = instancia_f
    independent_set = S 
    
    # Desenhar o grafo com o conjunto independente máximo destacado
    draw_graph_with_independent_set(graph, independent_set)
else:
    print("Não existe solução para a instância I do problema A'")


```
